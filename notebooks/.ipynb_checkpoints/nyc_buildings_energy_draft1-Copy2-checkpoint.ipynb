```python
import pandas as pd
import scipy.stats as stats
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.externals.six import StringIO
from sklearn.preprocessing import StandardScaler

#visualization
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image
import pydot
```

```python
pluto_mn = pd.read_csv("../pluto/MN.csv")
pluto_bx = pd.read_csv("../pluto/BX.csv")
pluto_bk = pd.read_csv("../pluto/BK.csv")
pluto_qn = pd.read_csv("../pluto/QN.csv")
pluto_si = pd.read_csv("../pluto/SI.csv")

frames = [pluto_mn, pluto_bx, pluto_bk, pluto_qn, pluto_si]
pluto_nyc = pd.concat(frames)
#pluto_nyc.to_csv('../pluto/pluto_nyc.csv')
pluto_nyc.shape
```

```python
building_column_names = ["record","BBL","bbl_status","coreported_bbl","nyc_bin","street_number","street","borough",
                "zip","covered_building","benchmark_status","site_eui","norm_site_eui","source_eui","norm_source_eui",
                "water_intensity","water_benchmark_eligible","water_method","energy_star","total_ghg","direct_ghg",
                "indirect_ghg","building_sf","dof_gsf","primary_use","num_buildings"]
building_features = building_column_names[1:]
building_target = building_column_names[0]
building_data = pd.read_csv("../data/2015_nyc_cy2014__ll84_disclosure_data.csv",header=0,names=building_column_names)
building_data.head()
building_data.shape
```

```python
building_data_primary = building_data.loc[building_data['bbl_status'] != 'Secondary']
building_data_primary.shape
```

```python
building_data_join = pd.merge(building_data_primary, pluto_nyc, how='inner', on='BBL')
print building_data_join.head()
building_data_join.shape
```

```python
building_data_join.dtypes
```

```python
convert_to_float = ["coreported_bbl","nyc_bin","street_number","site_eui","norm_site_eui","source_eui",
                    "norm_source_eui","energy_star","total_ghg","direct_ghg","indirect_ghg","building_sf"]
```

```python
building_data_join[convert_to_float] = building_data_join[convert_to_float].convert_objects(convert_numeric=True)
#building_data_join.head()
building_data_join.dtypes
```

```python
print building_data_join.isnull().head()
building_data_join.isnull().sum()
```

```python
building_data_columns = ["BBL","covered_building","benchmark_status","site_eui","norm_site_eui","source_eui",
                         "norm_source_eui","energy_star","total_ghg","direct_ghg","indirect_ghg","building_sf",
                         "dof_gsf","primary_use","num_buildings","Borough","Block","Lot","Address","ZoneDist1",
                         "BldgClass","LandUse","LotArea","BldgArea","NumBldgs","NumFloors","AssessLand","AssessTot",
                         "YearBuilt","BuiltFAR","XCoord","YCoord"]

building_data_columns = building_data_columns[:1]+building_data_columns[3:13]+building_data_columns[14:15]+building_data_columns[15:32]+building_data_columns[1:3]+building_data_columns[13:14]+building_data_columns[15:22]
print building_data_columns

```

```python
building_energy_data = building_data_join[building_data_columns]
target = building_energy_data['LandUse']
print building_energy_data.dtypes
building_energy_data.head()
```

```python
building_energy_data.isnull().sum()
```

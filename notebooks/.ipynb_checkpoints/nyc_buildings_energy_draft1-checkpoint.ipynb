```python
pluto_mn = pd.read_csv("../pluto/MN.csv")
pluto_bx = pd.read_csv("../pluto/BX.csv")
pluto_bk = pd.read_csv("../pluto/BK.csv")
pluto_qn = pd.read_csv("../pluto/QN.csv")
pluto_si = pd.read_csv("../pluto/SI.csv")

frames = [pluto_mn, pluto_bx, pluto_bk, pluto_qn, pluto_si]
pluto_nyc = pd.concat(frames)
#pluto_nyc.to_csv('../pluto/pluto_nyc.csv')
pluto_nyc.shape
```

```python
building_column_names = ["record","BBL","bbl_status","coreported_bbl","nyc_bin","street_number","street","borough",
                "zip","covered_building","benchmark_status","site_eui","norm_site_eui","source_eui","norm_source_eui",
                "water_intensity","water_benchmark_eligible","water_method","energy_star","total_ghg","direct_ghg",
                "indirect_ghg","building_sf","dof_gsf","primary_use","num_buildings"]
building_features = building_column_names[1:]
building_target = building_column_names[0]
building_data = pd.read_csv("../data/2015_nyc_cy2014__ll84_disclosure_data.csv",header=0,names=building_column_names)
building_data.head()
```

```python
building_data[convert_to_float] = building_data[convert_to_float].convert_objects(convert_numeric=True)
building_data.head()
building_data.dtypes
# building_data.apply(pd.to_numeric)
```

```python
building_data.isnull().head()
building_data.isnull().sum()
```

```python
building_data_columns = list(building_data.columns.values)
building_data_columns = building_data_columns[:2]+building_data_columns[4:5]+building_data_columns[11:15]+building_data_columns[18:24]+building_data_columns[-1:]+building_data_columns[2:4]+building_data_columns[5:11]+building_data_columns[15:18]+building_data_columns[-2:-1]
building_data_columns
```

```python
building_data = building_data[building_data_columns]
target = building_data['primary_use']
print building_data.dtypes
building_data.head()
```

```python
print building_data[building_data_columns[:14]].head(30)
median_per_column = building_data[building_data_columns[3:14]].apply(lambda x: x.median(),axis=0)
print "Median per column:\n",median_per_column
```

```python
numeric_median_filled = building_data[building_data_columns[3:14]].fillna(median_per_column,axis=0)
numeric_median_filled.head(30)
```

```python
def get_most_frequent_value(my_column):
    return my_column.value_counts().index[0]

most_frequent_values_per_column = building_data[building_data_columns[-1:]].apply(get_most_frequent_value,axis=0)
print "Most frequent value in each column:\n",most_frequent_values_per_column

categorical_most_frequent = building_data[building_data_columns[-1:]].fillna(most_frequent_values_per_column,axis=0)
categorical_most_frequent.head()
```

```python
building_data_numeric_columns = building_data.columns.tolist()[3:14]
building_data_categorical_columns = building_data.columns.tolist()[-1:]
building_data_filled = building_data.copy()
building_data_filled[building_data_numeric_columns] = numeric_median_filled
building_data_filled[building_data_categorical_columns] = categorical_most_frequent
building_data_filled.isnull().sum()
```

```python
#random forest regressor and check 10-fold RMSE
X_final = final_building_data.ix[:,:9]
y_target = final_building_data['total_ghg']

rf = RandomForestRegressor()
cross_val_scores = np.abs(cross_val_score(rf,X_final,y_target,scoring = "mean_squared_error", cv=10))
rmse_cross_val_scores = map(np.sqrt, cross_val_scores)
print "Mean 10-fold rmse: ", np.mean(rmse_cross_val_scores)
print "Std 10-fold rmse: ", np.std(rmse_cross_val_scores)
```

```python
# list of values to try for max_features
feature_names_buildings = list(final_building_data.ix[:,:9].columns.values)
feature_range = range(1, len(feature_names_buildings)+1)

X_buildings = final_building_data[feature_names_buildings]
y_buildings = final_building_data['total_ghg']
X_train,X_test,y_train,y_test = train_test_split(X_buildings,y_buildings,test_size=0.4)


# list to store the average RMSE for each value of max_features
RMSE_scores = []

for feature in feature_range:
    rfreg = RandomForestRegressor(n_estimators=220, bootstrap=True, 
                                  oob_score=True, max_features=feature, random_state=1234)
    rfreg.fit(X_train,y_train)
    preds = rfreg.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    RMSE_scores.append(rmse)
```

```python
#find number of estimaters that provide best prediction

# plot n_estimators (x-axis) versus RMSE (y-axis)
plt.plot(feature_range, RMSE_scores)
plt.xlabel('max_features')
plt.ylabel('RMSE (lower is better)')
```

```python
#another random forest regrossor with train-test-split to find feature importance

rfreg_best = RandomForestRegressor(n_estimators=180, max_features=7, bootstrap=True, oob_score=True, random_state=123)
rfreg_best.fit(X_train,y_train)
preds = rfreg_best.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test,preds))
print "Best RMSE:",rmse
```

```python
pd.DataFrame({'feature':feature_names_buildings, 
              'importance':rfreg_best.feature_importances_}).sort_values(by='importance',ascending=False)
```
